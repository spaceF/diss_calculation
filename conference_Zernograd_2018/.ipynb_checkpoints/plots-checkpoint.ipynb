```python
import pandas as pd
```

Construct the graphs of cost fuel

```python
data = pd.read_csv('fuel.csv', sep=";")
```

```python
data
```

```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
year = np.array(data['year'])
gasoline = np.array(data['gasoline'])
diesel = np.array(data['diesel'])
```

```python
sns.set(style="ticks")
```

```python
dpi =100
fig = plt.figure(dpi=dpi, figsize=(700 / dpi, 500 / dpi))

plt.rcParams.update({'font.size': 14})

plt.plot(year, gasoline, label='бензин')
plt.plot(year, diesel, label='дизель')
plt.xlabel("Год")
plt.ylabel("Стоимость, руб")
plt.legend(loc=2,prop={'size':15})
plt.grid(True, linestyle='-', color='0.75')
plt.show()

fig.savefig('plot_1.png')
```

```python
data = pd.read_csv('load.csv', sep=";")
```

```python
data
```

```python
data.loc[[0,1,2,3], ['P']] = 0
data.loc[[22,23], ['P']] = 0
data.loc[[7], ['P']] = 1.687
data.loc[[12], ['P']] = 2.456
data.loc[[13], ['P']] = 2.678
data.loc[[20], ['P']] = 1.345
data.loc[[21], ['P']] = 1.345
```

```python
data.loc[[0,1,2,3], ['P']]
```

```python
clock = np.array(data['clock'])
P = np.array(data['P'])
```

```python
dpi =100
fig = plt.figure(dpi=dpi, figsize=(700 / dpi, 500 / dpi))

plt.rcParams.update({'font.size': 15})

plt.bar(clock, P, align='center', width=1, edgecolor='black', color='green')
plt.xlabel("Часы")
plt.ylabel("Мощность, кВт")
plt.legend(loc=2,prop={'size':15})
plt.grid(True, linestyle='-', color='0.75')
plt.show()

#fig.savefig('load.png')
```
